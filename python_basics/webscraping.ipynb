<h1>Scrape data from HTML tables</h1>

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/HTMLColorCodes.html"
data = requests.get(url).text
soup = BeautifulSoup(data, "html.parser")
table = soup.find("table")
for row in table.find_all('tr'):
    col = row.find_all('td')
    color_name = col[2].string
    color_code = col[3].string
    if color_code is not None:
        print("{} --> {}".format(color_name, color_code))
    else:
        print("Color Name --> Color Code")
```

<h1>Scrape data from HTML tables into a DataFrame using BeautifulSoup and Pandas</h1>

```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/World_population"
data = requests.get(url).text
soup = BeautifulSoup(data, 'html.parser')
tables = soup.find_all('table')
len(tables)
```

```python
#looking for the 10 most densly populated countries table
for index, table in enumerate(tables):
    if ("10 most densely populated countries" in str(table)):
        table_index = index
print(table_index)
```

```python
#locate the above table
print(tables[table_index].prettify())
```

```python
#put data from table into a DataFrame
population_data = pd.DataFrame(columns=["Rank", "Country", "Population", "Area", "Density"])
for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if col != []:
        rank = col[0].text
        country = col[1].text
        pop = col[2].text.strip()
        area = col[3].text.strip()
        density = col[4].text.strip()
        population_data = population_data.append({"Rank": rank, "Country": country, "Population": pop, "Area": area, "Density": density}, ignore_index=True)
population_data
```

<h3>Scrape data from HTML tables into a DataFrame using BeautifulSoup and read_html</h3>

```python
population_data_html = pd.read_html(str(tables[table_index]), flavor="bs4")[0]
population_data_html
```
