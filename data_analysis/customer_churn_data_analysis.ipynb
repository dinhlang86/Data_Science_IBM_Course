<h3>Handle Missing Values</h3>

```python
# Detecting the missing values
# Using error = â€˜coerceâ€™ will replace all non-numeric values with NaN
df.TotalCharges = pd.to_numeric(df.TotalCharges, errors='coerce')
df.isnull().sum()
```

```python
# 11 missing values in TotalCharges is too small with the dataset, so we drop them
df.dropna(how='any', inplace=True)
df.isnull().sum()
```

<h3>Data Visualization</h3>

```python
# Visualize the target variable (Churn)
plt.figure(figsize=(5, 5))
sns.countplot(x=df['Churn'], palette='Blues', linewidth=1)
plt.show()
```

The plot shows a class imbalance of the data between churners and non-churners. To address this, resampling would be a suitable approach.

```python
for i, predictor in enumerate(df.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Churn')
```

Relationship between Monthly Charges and Total Charges:

```python
# Total Charges is the sum of Monthly Charges. 
sns.lmplot(data=df, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

The TotalCharges and MonthlyCharges are highly correlated

```python
# Customer Contract Distribution
sns.histplot(data=df, x='Churn', hue='Contract', multiple='dodge', shrink=0.8).set(title="Customer Contract Distribution")
```

About 75% of customers with Month-to-Month Contracts opted to move out as compared to 13% of customers with one-year contract and 3% with two-year contracts.

```python
# Payment Method Distribution
plt.pie(df['PaymentMethod'].value_counts(), labels=df['PaymentMethod'].unique(), autopct='%1.2f%%')
plt.show()
```

The electronic check has the highest users.

```python
# Dependents Distribution
sns.histplot(data=df, x='Churn', hue='Dependents', multiple='dodge', shrink=.8).set(title="Dependents Distribution")
```

Customers without dependents are more likely to churn

```python
# Partners Distribution
sns.histplot(data=df, x='Churn', hue='Partner', multiple='dodge', shrink=.8).set(title="Partners Distribution")
```

Customers that do not have partners are likely to churn more.

<h3>Customers with month-to-month contracts churn the most. 
Total charges and monthly charges were highly correlated</h3>
